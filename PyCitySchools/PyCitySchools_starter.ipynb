```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.to_csv('school_data_complete', index = False)
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = len(school_data)
total_students = len(student_data)
total_budget = school_data["budget"].sum()

avg_math = student_data["math_score"].mean()
avg_read = student_data["reading_score"].mean()

passing_math = student_data.apply(lambda x: True if x['math_score'] > 69 else False , axis = 1)
num_passing_math = len(passing_math[passing_math == True].index)
perc_passing_math = (num_passing_math / total_students) * 100

passing_read = student_data.apply(lambda x: True if x['reading_score'] > 69 else False , axis = 1)
num_passing_read = len(passing_read[passing_read == True].index)
perc_passing_read = (num_passing_read / total_students) * 100

overall_passing = (perc_passing_math + perc_passing_read)/2

district = {
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math],
    "Average Reading Score": [avg_read],
    "% Passing Math": [perc_passing_math],
    "% Passing Reading": [perc_passing_read],
    "% Overall Passing Rate": [overall_passing]
}


district_df = pd.DataFrame(district)
district_df
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
school_data.insert(5,"passing math",0)
school_data.insert(6,"passing reading",0)
school_data.insert(7,"overall passing",0)
school_data.insert(8,"# passing math",0)
school_data.insert(9,"# passing reading",0)
school_data.insert(10,"avg math",0)
school_data.insert(11,"avg reading",0)


passingm = 0
passingr = 0
index = 0
length = len(school_data_complete)
first = True
readtot = 0
mathtot = 0

for row in school_data_complete.itertuples():
    if first == True:
        last_name = row[5]
        size = row[10]
        first = False
        
    if row[5] == last_name:
        
        readtot += row[6]
        mathtot += row[7]
        
        if row[6] > 69:
            passingr = passingr + 1
            
        if row[7] > 69:
            passingm = passingm + 1
    
    if row[1] == length - 1:
        perc_pass_m = (passingm / size)*100
        perc_pass_r = (passingr / size)*100
        ovr_pass = (perc_pass_m + perc_pass_r)/2
        mathavg = mathtot / size
        readavg = readtot / size
        school_data.loc[index, 'passing math'] = perc_pass_m
        school_data.loc[index, 'passing reading'] = perc_pass_r
        school_data.loc[index, 'overall passing'] = ovr_pass
        school_data.loc[index, '# passing math'] = passingm
        school_data.loc[index, '# passing reading'] = passingr
        school_data.loc[index, 'avg math'] = mathavg
        school_data.loc[index, 'avg reading'] = readavg
            
    elif row[5] != last_name:
        perc_pass_m = (passingm / size) * 100
        perc_pass_r = (passingr / size) * 100
        ovr_pass = (perc_pass_m + perc_pass_r)/2
        mathavg = mathtot / size
        readavg = readtot / size
        school_data.loc[index, 'passing math'] = perc_pass_m
        school_data.loc[index, 'passing reading'] = perc_pass_r
        school_data.loc[index, 'overall passing'] = ovr_pass
        school_data.loc[index, '# passing math'] = passingm
        school_data.loc[index, '# passing reading'] = passingr
        school_data.loc[index, 'avg math'] = mathavg
        school_data.loc[index, 'avg reading'] = readavg
        index = index + 1
        passingm = 0
        passingr = 0
        readtot = 0
        mathtot = 0
        size = row[10]
        last_name = row[5]
        
next

school_data.to_csv('school_data', index = False)
top_sorted_df = school_data.sort_values(by=['overall passing'], ascending = False)
top_sorted_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
spending_df = school_data[['size', 'budget', 'avg math', 'avg reading', 'passing math', 'passing reading', 'overall passing',]].copy()
budget_per = school_data['budget'] / school_data['size']
spending_df['budget per student'] = budget_per

spending_df["budget group"] = pd.cut(spending_df["budget per student"], spending_bins, labels = group_names)

spending_group_df = spending_df.groupby("budget group")
spending_group_df['avg math', 'avg reading', 'passing math', 'passing reading', 'overall passing'].mean()


```
